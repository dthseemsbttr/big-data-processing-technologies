3.2 Разбейте каждый из имеющихся датасетов на две части: 1 часть содержит только те строки, где PDV (Percent Daily Value) превышает 100%; 2 часть содержит те строки, где PDV составляет не более 100%. Создайте 2 группы в файле и разместите в них соответствующие части датасета c сохранением метаданных исходных датасетов. Итого должно получиться 2 группы, содержащие несколько датасетов. Датасеты, которые не содержат информацию о PDV, оставьте вне групп. Сохраните результаты в файл `nutrition_grouped.h5`.

````python
with h5py.File(
    "nutrition_sample.h5",
    "r"
) as ns:

    ds = {}
    attrs = {}
    ds_0 = {}
    attrs_0 = {}
    for key in ns.keys():
        if 'PDV' in ns[key].attrs['col_1']:
            ds[key] = ns[key][:]
            attrs[key] = dict(ns[key].attrs)
        else:
            ds_0[key] = ns[key][:]
            attrs_0[key] = dict(ns[key].attrs)

pdv = {
    'more_than_100': {},
    'less_equal_than_100': {}
}

for key, value in ds.items():
    pdv['more_than_100'][key] = value[value[:,1] > 1]
    pdv['less_equal_than_100'][key] = value[value[:,1] <= 1]

with h5py.File(
    "nutrition_grouped.h5",
    "w"
) as f:
    dataset = f.create_dataset(name = 'dataset_0', data = ds_0['dataset_0'])
    dataset.attrs.update(attrs_0['dataset_0'])
    for group, data in pdv.items():
        new_group = f.create_group(group)
        for k, d in data.items():
            item = new_group.create_dataset(name=k, data=d)
            item.attrs.update(attrs[k])
````

3.3 Выведите названия всех групп и датасетов, находящихся в этих группах, из файла `nutrition_grouped.h5` а также размерность матриц, содержащихся в датасетах и их метаданные.

Формат вывода:
```
Dataset name=dataset_0 n_rows=30000 n_cols=2 col_0=recipe_id col_1=calories (#)
Group less_equal_than_100:
    Dataset name=less_equal_than_100/dataset_1 n_rows=28264 n_cols=2 col_0=recipe_id col_1=total fat (PDV)
    ....
Group more_than_100
    Dataset name=more_than_100/dataset_1 n_rows=1736 n_cols=2 col_0=recipe_id col_1=total fat (PDV)
    ....
....
```

````python
with h5py.File(
    "nutrition_grouped.h5",
    "r"
) as ng:
    
    for key in ng.keys():
        if type(ng[key]) == h5py._hl.dataset.Dataset:
            print(f'Dataset name={key} n_rows={ng[key].shape[0]} n_cols={ng[key].shape[1]}', end = ' ')
            for attr_key in ng[key].attrs.keys():
                print(f'{attr_key}={ng[key].attrs[attr_key]}', end = ' ')
            print()
        else:
            print(f'Group {key}:')
            for d in ng[key].keys():
                print(f'\tDataset name={d} n_rows={ng[key][d].shape[0]} n_cols={ng[key][d].shape[1]}', end = ' ')
                for attr_key in ng[key][d].attrs.keys():
                    print(f'{attr_key}={ng[key][d].attrs[attr_key]}', end = ' ')
                print()            
````
